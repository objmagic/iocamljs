### Let-bound polymorphism

One of the restrictions of type inference is that only let-bound variables, not lambda-bound variables, can have generalized types.  In the following example, `one` is given a polymorphic type (more strictly: a type scheme) and then used polymorphically in the definition of `p`.

```ocaml
let one x = [x]
let p = (one false, one "a")
```

However, if we try to make `one` a parameter to `p` OCaml rejects the definition since the types of parameters are not generalized:

```ocaml
let p one = (one false, one "a")
```

Here's another example.  The list append function '@' can be applied to any type of list:

```ocaml
let double2 ((l : int list), (r: string list)) = (l @ l, r @ r)
```

However, if we try to turn `@` into a parameter OCaml rejects the definition.  Only functions which are used at a single type in a definition can be turned into parameters.

```ocaml
let double2 (@) ((l : int list), (r: string list)) = (l @ l, r @ r)
```

### Monomorphic recursion

Functions over non-regular types typically involve *polymorphic recursion*.  The following type of balanced trees from Lecture 2 is non-regular because `perfect` is applied to `'a * 'a` rather than to `'a` in the type of `SuccP`:

```ocaml
type 'a perfect =
       ZeroP : 'a -> 'a perfect
     | SuccP : ('a * 'a) perfect -> 'a perfect
```

We can write a recursive function to compute the depth of the tree, but OCaml rejects the definition.  The type of `depthP` should be `'a perfect -> int`, but it is applied to `p`, which has type `('a * 'a) perfect`:

```ocaml
let rec depthP = function
       ZeroP _ -> 0
     | SuccP p -> succ (depthP p)
```

However, the definition is accepted if we supply a scheme for the function.  When we supply a scheme OCaml needs only to check the type, which is an easier problem than inferring a type.

```ocaml
let rec depthP : 'a. 'a perfect -> int = function
       ZeroP _ -> 0
     | SuccP p -> succ (depthP p)
```

### The value restriction

In the following definition, the type of `l` is generalised because `[[]]` is a syntactic value:

```ocaml
let l = [[]]
```

In the following definition, the type of `s` is not generalised because `ref []` is not a syntactic value.  OCaml indicates that the type variable `'_a'` is *weak* (i.e. ungeneralized) by prefixing it with an underscore.

```ocaml
let s = ref []
```

We can store a value of type `int list` to the cell:

```ocaml
s := [1; 2; 3]
```

Now if we ask for the type again we find that the weak type variable has been instantiated

```ocaml
s
```

#### Eta expansion

It is sometimes possible to recover polymorphism by eta expanding.  In the following definition `f` is not generalized because the right hand side of the definition is not a syntactic value:

```ocaml
let id x = x

let f  = id (fun _ -> "a")
```

However if we eta-expand the right hand side then the type is generalized:

```ocaml
let f = fun x -> id (fun _ -> "a") x
```

### Covariance and contravariance

OCaml provides syntax for annotating type parameters with variance.  A type parameter can be prefixed with `+` or `-` to mark it as covariant or contravariant.  OCaml will check that the parameter is only used with the variance that you specify.  For example, try changing `-` to `+` in the definition of `printer`:

```ocaml
type -'a printer = 'a -> string
```

```ocaml
type (+'a, -'b) t = ('a -> 'b) -> 'a printer -> 'b printer list
```

### The relaxed value restriction

```ocaml
let rec repeat n x =
   if n = 0 then []
   else x :: repeat (n - 1) x
```

In the following definition the type of `s` is generalized because its type variable appears in a covariant positions:

```ocaml
let s = repeat 3 []
```

We can even use imperative features without losing polymorphism.  Only the variance of the type variables is significant in deciding whether to generalize when the right hand side of a let binding is not a value:

```ocaml
let s =
   let x = ref [] in
      x := [[]; []; []];
      !x
```

In the following definition `s` is not generalized because the type variable appears in a contravariant position:

```ocaml
let id x = x
let s = id (fun _ -> "")
```

### The occurs check

The following definition is rejected because the parameter `g` is used at two different types.  Unifying these types would create a recursive type, which is prevented by the *occurs check* in the unification algorithm.

```ocaml
let f g = g g
```

We can disable the occurs check by enabling `rectypes`.  With `rectypes` enabled the definition is accepted and the type inferred is recursive:

```ocaml
#rectypes
```

```ocaml
let f g = g g
```
