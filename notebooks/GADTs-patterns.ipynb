## GADTs and equality

GADTs allow us to encode type equality more directly than in System F$\omega$.  We can define equality as a GADT `eql` with two type indexes and a single nullary constructor, `Refl`:

```ocaml
type (_, _) eql = Refl : ('a,'a) eql
```


Using `Refl` the standard equality properties can be encoded much more straightforwardly than in System F$\omega$.  Here are definitions of the symmetry and transitivity properties:

```ocaml
let symm : type a b. (a, b) eql -> (b, a) eql =
  fun Refl -> Refl
```

```ocaml
let trans : type a b c. (a, b) eql -> (b, c) eql -> (a, c) eql =
  fun Refl Refl -> Refl
```


We can also write a function which turns a proof that `a` and `b` are equal into a proof that `a t` and `a b` are equal for any type constructor `t`.  Since this function involves higher-kinded polymorphism we must make use of the module system:

```ocaml
module Lift(T : sig type _ t end) =
struct
  let lift : type a b. (a, b) eql -> (a T.t, b T.t) eql =
     fun Refl -> Refl
end
```


Finally, we can turn a proof of equality into a coercion.  If two types `a` and `b` are equal then it's safe to convert a value of type `a` into a value of type `b`:

```ocaml
let cast : type a b. (a, b) eql -> a -> b =
  fun Refl x -> x
```

```ocaml
type z = Z : z
type _ s = S : 'n -> 'n s
```

```ocaml
type ('a,'n) etree =
  EmptyE : (z,'n) eql  ->  ('a,'n) etree
| TreeE : ('n,'m s) eql *
   ('a,'m) etree * 'a * ('a,'m) etree  ->  ('a,'n) etree
```

```ocaml
let rec depthE : type a n.(a, n) etree  ->  n =
  function
    EmptyE Refl  ->  Z
  | TreeE (Refl, l,_,_)  ->  S (depthE l)
```


# GADT programming patterns

## *Pattern*: Building GADT values

It's not always possible to determine index types statically.  
For example, the depth of a tree might depend on user input.


#### Building GADT values using existentials

```ocaml
type (_,_,_) max =
    MaxEq : ('a,'a,'a) max
  | MaxFlip : ('a,'b,'c) max -> ('b,'a,'c) max
  | MaxSuc : ('a,'b,'a) max -> ('a s,'b,'a s) max
```

```ocaml
type ('a,_) dtree =
    EmptyD : ('a,z) dtree
  | TreeD : ('a,'m) dtree * 'a * ('a,'n) dtree * ('m,'n,'o) max 
    -> ('a,'o s) dtree
```

```ocaml
type 'a edtree = E : ('a, 'n) dtree -> 'a edtree
```


>let mk_dtree : type a n. a option -> a edtree =

```ocaml
function
  None -> E EmptyD
| Some v -> E (TreeD (EmptyD, v, EmptyD, MaxEq))
```


#### Building GADT values using universals

```ocaml
type ('a, 'k) adtree = { k: 'n. ('a, 'n) dtree -> 'k }
```

```ocaml
let mk_dtree_k: type a k.a option -> (a, k) adtree -> k =
fun opt {k} -> match opt with
  None -> k EmptyD
| Some v -> k (TreeD (EmptyD, v, EmptyD, MaxEq))
```



## *Pattern*: Singleton types

Without dependent types we can't write predicates involving data.  
Using one type per value allows us to simulate value indexing.

```ocaml
type z = Z
type 'n s = S : 'n -> 'n s
```

```ocaml
type (_,_,_) max =
  MaxEq : ('a,'a,'a) max
| MaxFlip : ('a,'b,'c) max -> ('b,'a,'c) max
| MaxSuc : ('a,'b,'a) max -> ('a s,'b,'a s) max
```

```ocaml
type (_,_,_) add =
    AddZ : (z,'n,'n) add
  | AddS : ('m,'n,'o) add  ->  ('m s,'n,'o s) add
```


## *Pattern*: Building evidence

With type refinement we learn about types by inspecting values.  
Predicates should return useful *evidence* rather than `true` or `false`.

```ocaml
type 'a tree =
  Empty : 'a tree
| Tree : 'a tree * 'a * 'a tree -> 'a tree
```

```ocaml
let top : 'a.'a tree -> 'a option =
  function
    Empty -> None
  | Tree (_,v,_) -> Some v
```

```ocaml
let is_empty : 'a .'a tree -> bool =
  function
    Empty  ->  true
  | Tree _  ->  false
```

```ocaml
let _ = fun t -> if not (is_empty t) then
  top t
else
  None
```


#### Evidence about whether a tree is empty

```ocaml
type _ is_zero =
  Is_zero : z is_zero
| Is_succ : _ s is_zero
```

```ocaml
type ('a,_) dtree =
  EmptyD : ('a,z) dtree
| TreeD : ('a,'m) dtree * 'a * ('a,'n) dtree * ('m,'n,'o) max 
      -> ('a,'o s) dtree
```

```ocaml
let is_empty : type a n.(a,n) dtree -> n is_zero =
  function
    EmptyD  ->  Is_zero
  | TreeD _  ->  Is_succ
```

```ocaml
let topD : type a n.(a,n s) dtree -> a =
  function TreeD (_,v,_,_) -> v
```

```ocaml
let _ = fun (type n) (t : (_, n) dtree) -> match is_empty t with
    Is_succ  ->  Some (topD t)
  | Is_zero  ->  None
```
