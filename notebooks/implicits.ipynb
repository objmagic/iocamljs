## Modular implicits

```ocaml
module type NUM = sig
  type t
  val zero : t
  val add : t -> t -> t
end
```

```ocaml
implicit module Num_int = struct
  type t = int
  let zero = 0
  let add x y = x + y
end
```

```ocaml
implicit module Num_float = struct
  type t = float
  let zero = 0.0
  let add x y = x +. y
end
```

```ocaml
let sum: {N:NUM} -> N.t list -> N.t =
 fun {N:NUM} l -> 
  fold_left N.add N.zero l 
```

```ocaml
sum [1;2;3]
sum [1.0;2.0;3.0]
```


### Implicit functors

```ocaml
 implicit module Num_pair{A:NUM}{B:NUM} = struct
   type t = A.t * B.t
   let zero = (A.zero, B.zero)
   let add (a1, b1) (a2, b2) =
     (A1.add a1 a2, B1.add b1 b2)
 end
```

```ocaml
 sum [(10, 1.0); (20, 2.0)]
```


### Inheritance

```ocaml
 module type FRACTIONAL = sig
   type t
   module Num : NUM with type t = t
   val div : t -> t -> t
 end
```

```ocaml
 implicit module Fractional_int = struct
   type t = int
   module Num = Num_int
   let div n d = n / d
 end
```

```ocaml
 implicit module Num_fractional{F:FRACTIONAL} = F.Num
```


### Ambiguity

```ocaml
module type SHOW =
  sig type t val show : t -> string end

implicit module Show_bool =
  struct type t = bool let show = string_of_bool end

implicit module Show_int =
  struct type t = int let show = string_of_int end

let show {S:SHOW} (x: S.t) = S.show x
let print x = show x
```


### Higher kinds

```ocaml
module type FUNCTOR = sig
  type +'a t
  val fmap : ('a -> 'b) -> 'a t -> 'b t
end
```

```ocaml
let fmap {F:FUNCTOR} f x = F.fmap 
let replace {F:FUNCTOR} x c = fmap (fun _ -> x) c
```

```ocaml
implicit module Functor_list = struct
  type 'a t = 'a list
  let fmap = List.map
end
```

```ocaml
implicit module Functor_option = struct
  type 'a t = 'a option
  let fmap f = function
   | None -> None
   | Some v -> Some (f v)
end
```
