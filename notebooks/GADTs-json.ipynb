## GADTs and equality

GADTs allow us to encode type equality more directly than in System F$\omega$.  We can define equality as a GADT `eql` with two type indexes and a single nullary constructor, `Refl`:

````ocaml
type (_, _) eql = Refl : ('a,'a) eql
````

Using `Refl` the standard equality properties can be encoded much more straightforwardly than in System F$\omega$.  Here are definitions of the symmetry and transitivity properties:

````ocaml
let symm : type a b. (a, b) eql -> (b, a) eql =
  fun Refl -> Refl
````

````ocaml
let trans : type a b c. (a, b) eql -> (b, c) eql -> (a, c) eql =
  fun Refl Refl -> Refl
````

We can also write a function which turns a proof that `a` and `b` are equal into a proof that `a t` and `a b` are equal for any type constructor `t`.  Since this function involves higher-kinded polymorphism we must make use of the module system:

````ocaml
module Lift(T : sig type _ t end) =
struct
  let lift : type a b. (a, b) eql -> (a T.t, b T.t) eql =
     fun Refl -> Refl
end
````

Finally, we can turn a proof of equality into a coercion.  If two types `a` and `b` are equal then it's safe to convert a value of type `a` into a value of type `b`:

````ocaml
let cast : type a b. (a, b) eql -> a -> b =
  fun Refl x -> x
````

````ocaml
type z = Z : z
type _ s = S : 'n -> 'n s
````

````ocaml
type ('a,'n) etree =
  EmptyE : (z,'n) eql  ->  ('a,'n) etree
| TreeE : ('n,'m s) eql *
   ('a,'m) etree * 'a * ('a,'m) etree  ->  ('a,'n) etree
````

````ocaml
let rec depthE : type a n.(a, n) etree  ->  n =
  function
    EmptyE Refl  ->  Z
  | TreeE (Refl, l,_,_)  ->  S (depthE l)
````

# GADT programming patterns

## *Example*: representing JSON

The following grammar represents a subset of [JSON](http://json.org) expressions:

```
json ::= string         json-seq ::= json
         number                      json , json-seq
         true
         false
         null
         []
         [ json-seq ]
            
```

Here's a JSON expression accepted by the grammar:

```
  [ "one", true, 3.4, [[ "four" ], [null]]]
```

We can represent JSON values in OCaml with an algebraic data type:

````ocaml
type ujson =
    UStr : string  ->  ujson
  | UNum : float  ->  ujson
  | UBool : bool  ->  ujson
  | UNull : ujson
  | UArr: ujson list  ->  ujson
````

JSON expressions can be straightforwardly translated into OCaml values:

````ocaml
let uj = UArr [UStr "one"; UBool true; UNum 3.4;
               UArr [UArr [UStr "four"]; UArr [UNull]]]
````

## *Pattern*: Richly typed data

Data may have finer structure than algebraic data types can express.  
GADT indexes allow us to specify constraints more precisely.

````ocaml
type _ tjson =
   Str : string  ->  string tjson
 | Num : float  ->  float tjson
 | Bool : bool  ->  bool tjson
 | Null : unit tjson
 | Arr : 'a tarr  ->  'a tjson
and _ tarr =
   Nil : unit tarr
 | :: : 'a tjson * 'b tarr  ->  ('a*'b) tarr
````

````ocaml
let tj = Arr (Str "one" :: Bool true :: Num 3.4 ::
     Arr (Arr (Str "four" :: Nil) :: Null :: Nil) :: Nil)
````

````ocaml
(* negate all the bools in a JSON value *)
let rec negate : type a.a tjson  ->  a tjson = function
   Bool true  ->  Bool false
 | Bool false  ->  Bool true
 | Arr arr  ->  Arr (negate_arr arr)
 | v  ->  v
and negate_arr : type a.a tarr  ->  a tarr = function
   Nil  ->  Nil
 | j :: js  ->  negate j :: negate_arr js
````

## *Pattern*: Building GADT values

It's not always possible to determine index types statically.  
For example, the depth of a tree might depend on user input.


#### Building GADT values using existentials

````ocaml
type etjson = ETJson : 'a tjson  ->  etjson
type etarr = ETArr : 'a tarr  ->  etarr
````

````ocaml
let rec tjson_of_ujson : ujson  ->  etjson = function
    UStr s  ->  ETJson (Str s)
  | UNum u  ->  ETJson (Num u)
  | UBool b  ->  ETJson (Bool b)
  | UNull  ->  ETJson Null
  | UArr arr  ->
    let ETArr arr' = tarr_of_uarr arr in
    ETJson (Arr arr')
and tarr_of_uarr : ujson list  ->  etarr = function
    []  ->  ETArr Nil
  | j :: js  -> 
    let ETJson j' = tjson_of_ujson j in
    let ETArr js' = tarr_of_uarr js in
    ETArr (j' :: js')
````

#### Building GADT values using universals

````ocaml
type 'k atjson = {k: 'a. 'a tjson  ->  'k}
type 'k atarr = {k: 'a. 'a tarr  ->  'k}
````

````ocaml
let rec tjson_of_ujson : ujson  ->  'k atjson  ->  'k =
  fun j {k=return}  ->  match j with
    UStr s  ->  return (Str s)
  | UNum u  ->  return (Num u)
  | UBool b  ->  return (Bool b)
  | UNull  ->  return Null
  | UArr arr  ->
    tarr_of_uarr arr {k = fun arr'  -> 
    return (Arr arr') }
and tarr_of_uarr : ujson list  ->  'k atarr  ->  'k =
  fun jl {k=return}  ->  match jl with
    []  ->  return Nil
  | j :: js  -> 
    tjson_of_ujson j {k = fun j'  -> 
    tarr_of_uarr js {k = fun js'  -> 
    return (j' :: js') }}
````

## *Pattern*: Singleton types

Without dependent types we can't write predicates involving data.  
Using one type per value allows us to simulate value indexing.


````ocaml
type z = Z
type 'n s = S : 'n -> 'n s
````

````ocaml
type (_,_,_) max =
    MaxEq : ('a,'b) eql -> ('a,'b,'a) max
  | MaxFlip : ('a,'b,'c) max -> ('b,'a,'c) max
  | MaxSuc : ('a,'b,'a) max -> ('a s,'b,'a s) max
````

````ocaml
type (_,_,_) add =
    AddZ : (z,'n,'n) add
  | AddS : ('m,'n,'o) add  ->  ('m s,'n,'o s) add
````

## *Pattern*: Separating types and data


Entangling proofs and data leads to redundant, inefficient code.  
Separate proofs make data reusable and help avoid slow traversals.


````ocaml
type _ tyjson =
    TyStr : string tyjson
  | TyNum : float tyjson
  | TyBool : bool tyjson
  | TyNull : 'a tyjson  ->  'a option tyjson
  | TyArr: 'a tyarr  ->  'a tyjson
and _ tyarr =
    TyNil : unit tyarr
  | ::    : 'a tyjson * 'b tyarr  ->  ('a * 'b) tyarr
````

````ocaml
let tyj = TyArr (TyStr :: TyBool :: TyNum ::
          TyArr (TyArr (TyStr :: TyNil)
                  :: TyNull TyBool :: TyNil) :: TyNil)
let j = ("one", (true, (3.4, (((("four", ()), (None, ())), ())))))
````

````ocaml
let rec negate : type a.a tjson  ->  a tjson = function
   Bool true  ->  Bool false
 | Bool false  ->  Bool true
 | Arr arr  ->  Arr (negate_arr arr)
 | v  ->  v
and negate_arr : type a.a tarr  ->  a tarr = function
   Nil  ->  Nil
 | j :: js  ->  negate j :: negate_arr js
````

````ocaml
let rec negateD : type a.a tyjson  ->  a  ->  a =
  fun t v -> match t, v with
   TyBool, true  ->  false
 | TyBool, false  -> true
 | TyArr a, arr  ->  (negate_arrD a arr)
 | _, v  ->  v
and negate_arrD : type a.a tyarr  ->  a  ->  a =
  fun t v -> match t, v with
   TyNil, ()  ->  ()
 | j :: js, (a, b)  ->  (negateD j a, negate_arrD js b)
````

````ocaml
let rec negateDS : type a.a tyjson  ->  a  ->  a = function
   TyBool  ->  (function false  ->  true
                   |     true  ->  false)
 | TyArr a  ->  negate_arrDS a
 | _  ->  (fun v  ->  v)
and negate_arrDS : type a.a tyarr  ->  a  ->  a = function
   TyNil  ->  (fun ()  ->  ()) 
 | j :: js  ->  let n = negateDS j
             and ns = negate_arrDS js in
                 (fun (a, b)  ->  (n a, ns b))
````

````ocaml
let rec unpack_ujson :
  type a.a tyjson  ->  ujson  ->  a option =
  fun ty v  ->  match ty, v with
    TyStr, UStr s  ->  Some s
  | TyNum, UNum u  ->  Some u
  | TyBool, UBool b  ->  Some b
  | TyNull _, UNull  ->  Some None
  | TyNull j, v  ->  (match unpack_ujson j v with
                        Some v  ->  Some (Some v)
                      | None  ->  None)
  | TyArr a, UArr arr  ->  unpack_uarr a arr
  | _ -> None
and unpack_uarr :
  type a.a tyarr  ->  ujson list  ->  a option =
  fun ty v -> match ty, v with
    TyNil, []  ->  Some ()
  | j :: js, v :: vs  -> 
      (match unpack_ujson j v, unpack_uarr js vs with
         Some v', Some vs'  ->  Some (v', vs')
       | _  ->  None)
  | _ -> None
````

## *Pattern*: Building evidence

With type refinement we learn about types by inspecting values.  
Predicates should return useful *evidence* rather than `true` or `false`.

````ocaml
type 'a tree =
  Empty : 'a tree
| Tree : 'a tree * 'a * 'a tree -> 'a tree
````

````ocaml
let top : 'a.'a tree -> 'a option =
  function
    Empty -> None
  | Tree (_,v,_) -> Some v
````

````ocaml
let is_empty : 'a .'a tree -> bool =
  function
    Empty  ->  true
  | Tree _  ->  false
````

````ocaml
let _ = fun t -> if not (is_empty t) then
  top t
else
  None
````

#### Evidence about whether a tree is empty

````ocaml
type _ is_zero =
  Is_zero : z is_zero
| Is_succ : _ s is_zero
````

````ocaml
type ('a,_) dtree =
  EmptyD : ('a,z) dtree
| TreeD : ('a,'m) dtree * 'a * ('a,'n) dtree * ('m,'n,'o) max 
      -> ('a,'o s) dtree
````

````ocaml
let is_empty : type a n.(a,n) dtree -> n is_zero =
  function
    EmptyD  ->  Is_zero
  | TreeD _  ->  Is_succ
````

````ocaml
let topD : type a n.(a,n s) dtree -> a =
  function TreeD (_,v,_,_) -> v
````

````ocaml
let _ = fun (type n) (t : (_, n) dtree) -> match is_empty t with
    Is_succ  ->  Some (topD t)
  | Is_zero  ->  None
````

#### Evidence about whether a JSON value is built only from strings and arrays

````ocaml
type _ str_tyjs =
    SStr : string str_tyjs
  | SArr : 'a str_tyarr  ->  'a str_tyjs
and 'a str_tyarr =
    SNil : unit str_tyarr
  | :: : 'a str_tyjs * 'b str_tyarr -> ('a*'b) str_tyarr
````

````ocaml
let rec is_stringy : type a.a tyjson -> a str_tyjs option =
 function
    TyStr  ->  Some SStr
  | TyNum  ->  None
  | TyBool  ->  None
  | TyNull _  ->  None
  | TyArr arr  ->  match is_stringy_array arr with
                None  ->  None
              | Some sarr  ->  Some (SArr sarr)
and is_stringy_array :
 type a.a tyarr  ->  a str_tyarr option =
 function
    TyNil  ->  Some SNil
  | x :: xs  -> 
    match is_stringy x, is_stringy_array xs with
        Some x, Some xs  ->  Some (x :: xs)
      | _  ->  None
````

````ocaml
let rec is_stringyV : type a.a tjson -> a str_tyjs option =
 function
    Str _  ->  Some SStr
  | Num _  ->  None
  | Bool _  ->  None
  | Null  ->  None
  | Arr arr  ->  match is_stringy_arrayV arr with
                None  ->  None
              | Some sarr  ->  Some (SArr sarr)
and is_stringy_arrayV :
 type a.a tarr  ->  a str_tyarr option =
 function
    Nil  ->  Some SNil
  | x :: xs  -> 
    match is_stringyV x, is_stringy_arrayV xs with
        Some x, Some xs  ->  Some (x :: xs)
      | _  ->  None
````
