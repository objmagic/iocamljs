## Monads etc.

### What do monads give us?

* A general approach to implementing custom effects

* A reusable interface to computation

* A way to structure effectful programs in a functional language

## Effects

An **effect** is anything a function does besides mapping inputs to outputs.

If an expression `M` evaluates to a value `V` and changing

```
let x = M in N
```

```
let x = V in N
```

changes the behaviour then `M` also performs effects.

### Example effects

** Effects available in OCaml**

* (higher-order) state

  `r := f; !r ()`

* exceptions

  `raise Not_found`

* I/O of various sorts

  `input_byte stdin`

* concurrency (interleaving)

  `Gc.finalise v f`

* non-termination

  `let rec f x = f x`

**Effects unavailable in OCaml**

* non-determinism

  `amb f g h`

* first-class continuations

  `escape x in e`

* polymorphic state

  `r := "one"; r := 2`

* checked exceptions

  `int -{IOError}-> bool`

### Capturing effects in the types

Some languages capture effects in the type system.

We could have two function arrows:

* a **pure** arrow `a -> b`

* an **effectful** arrow (or family of arrows) `a ~E~> b` annotated with the type `E` of effects the function performs.

An alternative: decompose effectful arrows into functions and computations

`a ~E~> b` becomes `a -> T b`

## Monads

Monads take this second approach, adding a type `'a t` of *computations* that
return a result of type `'a`.  There's a separate computation type for each
type of effect.

Here's a fragment of OCaml code that performs some effects, namely reading and
writing to a single reference cell:

```
let id = !counter in
let () = counter := id + 1 in
  string_of_int id
```

Here's a second program, written using monads, that also reads and writes to a
single reference:

```
get          >>= fun id ->
put (id + 1) >>= fun () ->
  return (string_of_int id)
```

Monads have a simple signature, with a type of computations and functions for
turning a value into a computation and for sequencing computations:

````ocaml
module type MONAD =
sig
 type 'a t
 val return : 'a -> 'a t
 val (>>=)  : 'a t -> ('a -> 'b t) -> 'b t
end
````


Implementations of the `MONAD` signature must satisfy three laws: 

```
return v >>= k  ≡  k v

v >>= return  ≡  v

(m >>= f) >>= g  ≡  m >>= (fun x -> f x >>= g)
```

### Example: a state monad

Each monad implementation typically extends the signature with additional
operations.  Here's a signature for dealing with mutable state, which adds
`get` and `put` functions for reading and writing the state, and a `runState`
function for actually running computations.

````ocaml
module type STATE = sig
  type state
  include MONAD
  val get : state t
  val put : state -> unit t
  val runState : 'a t -> init:state -> state * 'a
end
````


Here's an implementation of `STATE`, parameterised by the type of the state:

````ocaml
module State (S : sig type t end)
  : STATE with type state = S.t = struct
  type state = S.t
  type 'a t = state -> state * 'a
  let return v s = (s, v)
  let (>>=) m k s = let s', a = m s in k a s'
  let get s = (s, s)
  let put s' _ = (s', ())
  let runState m ~init = m init
end
````


We can use the state monad to write stateful computations, such as a tree
traversal that generates a fresh name for each leaf.  Here's a data type for
trees:

````ocaml
type 'a tree =
    Empty : 'a tree
  | Tree : 'a tree * 'a * 'a tree -> 'a tree
````


We can instantiate the state monad so that the state type is `int`.  The `int`
cell will serve as a counter for generating fresh names.

````ocaml
module IState = State (struct type t = int end)
````


We can write a `fresh_name` function which retrieves the value of the counter,
increments it, and uses the (original) value to generate a fresh name:

````ocaml
let fresh_name : string IState.t =
  let open IState in
  get         >>= fun i ->
  put (i + 1) >>= fun () ->
  return (Printf.sprintf "x%d" i)
````


The `label_tree` function constructs a computation which labels a tree:

````ocaml
let rec label_tree : 'a tree -> string tree IState.t =
  let open IState in function
    Empty -> return Empty
  | Tree (l, v, r) ->
    label_tree l >>= fun l ->
    fresh_name   >>= fun name ->
    label_tree r >>= fun r ->
    return (Tree (l, name, r))
````


Here's a computation which labels a tree:

````ocaml
let unlabeled = label_tree (Tree (Tree (Empty, 3, Empty), 5, Tree (Empty, 4, Tree (Empty, 2, Empty))))
````


Running the computation returns the final state together with the labeled tree:

````ocaml
IState.runState unlabeled ~init:0
````


### State satisfies the monad laws

```
  return v >>= k
≡ (definition of return, >>=)
  fun s -> let s', a = (fun s -> (s, v)) s in k a s'
≡ (β)
  fun s -> let s', a = (s, v) in k a s'
≡ (β for let)
  fun s -> k v s
≡ (η)
  k v
```

### Example: an exception monad

Let's look at a second monad implementation, for computations that can raise
exceptions.  The `ERROR` signature extends `MONAD` with an operation `raise`
for raising an exception and a function `_try_` for actually running
computations and handling any exceptions raised:

````ocaml
module type ERROR = sig
  type error
  include MONAD
  val raise : error -> 'a t
  val _try_ : 'a t -> catch:(error -> 'a) -> 'a
end
````


Here's an implementation of `ERROR`, parameterised by the type of exceptions:

````ocaml
module Error (E: sig type t end)
  : ERROR with type error = E.t = struct
  type error = E.t
  type 'a t =
     Val : 'a -> 'a t
   | Exn : error -> 'a t
  let return v = Val v
  let (>>=) m k = match m with
    Val v -> k v | Exn e -> Exn e
  let raise e = Exn e
  let _try_ m ~catch = match m with
    Val v -> v | Exn e -> catch e
end
````


Let's instantiate `Error` with a simple type of errors:

````ocaml
type zero = Zero : zero
module Error_zero = Error(struct type t = zero end)
````


We can use the instantiated error monad to write computations.  First, here's
a function that builds a computation to traverse a tree:

````ocaml
let rec mapMTree f = Error_zero.(function
   Empty -> return Empty
 | Tree (l, v, r) ->
   mapMTree f l >>= fun l ->
   f v          >>= fun v ->
   mapMTree f r >>= fun r ->
   return (Tree (l, v, r)))
````


Here's a call to `mapMTree` with a function that raises `Zero` whenever a node
is found to be `0`:

````ocaml
let check_nonzero =
  Error_zero.(
    mapMTree
    (fun v ->
       if v = 0 then raise Zero
       else return v))
````


Here's the error monad in action:

````ocaml
Error_zero.(
 _try_
   (check_nonzero (Tree (Tree (Empty, 3, Empty), 4, Tree (Empty, 0, Empty))) >>= fun _ ->
    return true)
 ~catch:(fun Zero -> false))
````

````ocaml
Error_zero.(
 _try_
   (check_nonzero (Tree (Tree (Empty, 3, Empty), 4, Tree (Empty, 1, Empty))) >>= fun _ ->
    return true)
 ~catch:(fun Zero -> false))
````



### Exception satisfies the monad laws

```
  v >>= return
≡ (definition of return, >>=)
  match v with Val v -> Val v | Exn e -> Exn e
≡ (η for sums)
  v
```

### Indexed monads

Indexed monads add an additional type parameter to `t` representing the type
of effects.

A computation of type `('e, 'a) t` *performs an effect* `'e` and *produces a
result* of type `'a`.

Here's the indexed monad signature:

````ocaml
module type INDEXED_MONAD =
sig
  type ('e, 'a) t
  val return : 'a -> (_, 'a) t
  val (>>=) : ('e,'a) t ->
       ('a -> ('e,'b) t) ->
              ('e,'b) t
end
````


### An indexed monad for exceptions

Here's an indexed monad version of the `ERROR` signature.  Note that the
`error` type is no longer needed in the signature, since `t` is now
parameterised by the type of errors:

````ocaml
module type IERROR =
sig
  include INDEXED_MONAD
  val raise : 'e -> ('e, _) t
  val _try_ : ('e, 'a) t ->
        catch:('e -> 'a) ->
              'a
end
````


Here's an implementation of `IERROR`:

````ocaml
module IError : IERROR =
struct
  type ('e, 'a) t =
      Val : 'a -> ('e, 'a) t
    | Exn : 'e -> ('e, 'a) t
  let return v = Val v
  let raise e = Exn e
  let (>>=) m k =
    match m with
      Val v -> k v
    | Exn e -> Exn e
  let _try_ m ~catch =
    match m with
      Val v -> v
    | Exn e -> catch e 
end
````


Indexed monads work well with rows.  Here's a function for constructing a
computation that searches a list for a value satisfying a predicate `p` and
raising `Not_found` if there is no such element.

````ocaml
let rec find p = IError.(function
  [] -> raise `Not_found
| x :: _ when p x -> return x
| _ :: xs -> find p xs)
````


Note the type of `find`, which shows what errors the computation may raise.

Here's a second function for constructing computations which return the head
and tail of a list, raise `Empty` if the list is empty:

````ocaml
let pop = IError.(function
  [] -> raise (`Empty "pop")
| x :: xs -> return (x, xs))
````


Again, the type of `pop` shows what errors the computation may raise.

We can combine `pop` and `find` in a single computation:

````ocaml
let gt_0 x = x > 0

open IError
````

>

````ocaml
let result = 
  pop []       >>= fun (_, xs) ->
  find gt_0 xs >>= fun y ->
  return y
````


The type of the computation shows that it may raise either `Not_found` or `Empty`.

Here's a version of the above code written without monads:

````ocaml
open Pervasives
 
let rec find p = function
  [] -> raise Not_found
| x :: _ when p x -> x
| _ :: xs -> find p xs
 
exception Empty
 
let pop = function
  [] -> raise Empty
| x :: xs -> (x, xs)
 
let gt_0 x = x > 0
  
let result =
  let _, xs = pop [] in
  let y = find gt_0 xs in
    y
````


With this code the types say nothing about the exceptions that may be raised.

### Parameterised monads and Hoare Logic

Parameterised monads add two additional type parameters to `t` representing
the start and end states of a computation.

A computation of type `('p, 'q, 'a) t` has *precondition* (or starting state)
`'p` and *postcondition* (or ending state) `'q` and *produces a result* of
type `'a`.

i.e. `('p, 'q, 'a) t` is a kind of Hoare triple `{P} M {Q}`.

Here's the parameterised monad signature:

````ocaml
module type PARAMETERISED_MONAD =
sig
  type ('s,'t,'a) t
  val return : 'a -> ('s,'s,'a) t
  val (>>=) : ('r,'s,'a) t ->
       ('a -> ('s,'t,'b) t) ->
              ('r,'t,'b) t
end
````


Here's a parameterised monad version of the `STATE` signature, using the extra
parameters to represent the type of the reference cell.

````ocaml
module type PSTATE =
sig
 include PARAMETERISED_MONAD
 val get : ('s,'s,'s) t
 val put : 's -> (_,'s,unit) t
 val runState : ('s,'t,'a) t -> init:'s -> 't * 'a
end
````



The types show how the cell can change type during the computation --- for
example, the type of `bind`

```
   val (>>=) : ('r,'s,'a) t ->
        ('a -> ('s,'t,'b) t) ->
               ('r,'t,'b) t
```

says that

* the first argument is a computation that changes the type of the cell from
  `'r` to `'s`

* the second argument constructs a computation that changes the type of the
  cell from `'s` to `'t`,

* the result is a computation that changes the type of the
  cell from `'r` to `'t`,

Here's an implementation of `PSTATE`.

````ocaml
module PState : PSTATE =
struct
  type ('s, 't, 'a) t = 's -> 't * 'a
  let return v s = (s, v)
  let (>>=) m k s = let t, a = m s in k a t
  let put s _ = (s, ())
  let get s = (s, s)
  let runState m ~init = m init
end
````


### Programming with polymorphic state

We'll build a tiny typed stack machine with three operations.  Here's a
signature for the operations:

````ocaml
module type STACK_OPS =
sig
  type ('s,'t,'a) t
  val add : unit -> (int * (int * 's),
                            int * 's, unit) t
  val _if_ : unit -> (bool * ('a * ('a * 's)),
                                    'a * 's, unit) t
  val push_const : 'a -> ('s,
                     'a * 's, unit) t
end
````


The type of each operations show how that operation changes the stack.  For
example, for the `add` operation,

* the stack starts with a type of `int * (int * 's)` --- i.e. two integers on
  top, followed by some other values `'s`

* the stack ends with a type of `int * 's` --- i.e. one integer on
  top, followed by the other values `'s`

We can combine the stack operations with the parameterised monad signature to
build a signature for a stack machine:

````ocaml
module type STACKM = sig
 include PARAMETERISED_MONAD
 include STACK_OPS
   with type ('s,'t,'a) t := ('s,'t,'a) t
 val execute : ('s,'t,'a) t -> 's -> 't * 'a
end
````


Here's an implementation of the stack machine

````ocaml
module StackM : STACKM =
struct
  include PState
 
  let add () =
    get >>= fun (x,(y,s)) ->
    put (x+y,s)
 
  let _if_ () =
    get >>= fun (c,(t,(e,s))) ->
    put ((if c then t else e),s)

  let push_const k =
    get >>= fun s ->
    put (k, s)

  let execute c s = runState ~init:s c
end
````


Here's a computation built with the `StackM` monad:

````ocaml
let program = let open StackM in
  push_const 3 >>= fun () ->
  push_const 4 >>= fun () ->
  push_const 5 >>= fun () ->
  push_const true >>= fun () ->
  _if_ () >>= fun () ->
  add ()
````


We can run the computation to transform a stack:

````ocaml
StackM.execute program (10, (20, ()))
````
