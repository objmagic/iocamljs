## Generic programming

Generic programming involves functions which operate over a wide range
of data types.  Unlike (parametrically) polymorphic functions, which
operate uniformly at all types, generic functions behave in a way that
varies predictably for each type.

For any type `t` from variants, records and numeric types we can define a
variety of functions, including

* **equality**

  ```
  t -> t -> bool
  ```

* **ordering**

  ```
  t -> t -> int
  ```

* **hashing**

  ```
  t -> int
  ```

* **pretty-printing**

  ```
  t -> string
  ```

* **parsing**

  ```
  string -> t
  ```

* **serialising**

  ```
  t -> string
  ```

* **sizing**

  ```
  t -> int
  ```

OCaml provides a number of built-in generic functions, including
functions for equality, hashing, and serialisation.  Unfortunately, as
we have seen, these functions are not compatible with parametricity.
We can write parametric generic functions that operate on a wide
variety of data by adding an additional parameter that describes the
*shape* of the data.  For example, OCaml's built-in polymorphic
equality has the following type:

```
'a -> 'a -> bool
```

Our parametric equality function will have an additional implicit
argument of type `DATA` that describes the shape of the values to be
compared for equality:

```
{D:DATA} -> D.t -> D.t -> bool
```

Most of this notebook is about one approach to defining and using the
`DATA` signature.

## Type-indexed values

`DATA` is an example of using modules to define *type-indexed values*.  The
idea of type-indexed values is to represent types by defining a signature

```
module type T =
sig
  type t
  ...
end
```

together with an (implicit) instance of the signature for each type
constructor in the program

```
implicit module Bool : T with type t = bool
implicit module Int : T with type t = int
implicit module String : T with type t = string
implicit module Pair {A:T} {B:T} : T with type t = A.t * B.t
implicit module Option {A:T} : T with type t = A.t option
implicit module List {A:T} : T with type t = A.t list
(* etc. *)
```

and then using these instances to build representations of type
expressions.  For example, we might represent the type

```
(int * bool) option list
```

using the module expression:

```
List{Option{Pair{Int}{bool}}}
```

Type-indexed values can be used to define a variety of generic
functions.  For example, we might define a type-indexed value that
allows us to compare values for equality:

````ocaml
module type EQ = sig
  type t
  val eq: t -> t -> bool
end
implicit module Eq_bool : EQ with type t = bool = struct
  type t = bool
  let eq l r = match l, r with
   | false, false
   | true , true -> true
   | _           -> false
end
implicit module Eq_pair{A:EQ} {B:EQ} : EQ with type t = A.t * B.t = struct
  type t = A.t * B.t
  let eq (l1, r1) (l2, r2) = A.eq l1 l2 && B.eq r1 r2
end

(* etc. *)
````


or for printing

````ocaml
module type SHOW = sig
  type t
  val show : t -> string
end
implicit module Show_bool : SHOW with type t = bool = struct
  type t = bool
  let show = function
   | false -> "false"
   | true  -> "true"
end
implicit module Show_pair{A:SHOW} {B:SHOW}
 : SHOW with type t = A.t * B.t =
struct
  type t = A.t * B.t
  let show (l, r) =  "("^ A.show l ^", "^ B.show r ^")"
end

(* etc. *)
````


We're going to look at defining more generic type-indexed values that
make it possible to write a wide variety of generic functions in a
concise way.  In particular, we'll define:

* a type-indexed value for testing whether two types are equal
* a type-indexed value for accessing the immediate sub-nodes of a value.

These two type-indexed values will form the basis of a generic
programming library.

## Type equality

We'll start by building a type-indexed value for type equality --- that is,
for testing whether two types are equal.  We'll need a representation type,
`typerep`, to represent types as concrete values:

```
type 'a typerep
```

Besides a value of type `typerep`, the `TYPEABLE` signature should expose an
operation, `eqty`, which compares two `typerep` values for equality, returning
a GADT as proof if they're determined to be equal:

```
module type TYPEABLE = sig
  type t
  val typerep : t typerep
  val eqrep : 'other typerep -> (t, 'other) eql option
end

```

The `eql` type is the familiar equality GADT:

````ocaml
type (_, _) eql = Refl : ('a, 'a) eql
````


Finally, we'll need an instance of `TYPEABLE` for each type constructor:

```
implicit module Typeable_int : TYPEABLE with type t = int
implicit module Typeable_bool : TYPEABLE with type t = bool
implicit module Typeable_list{A:TYPEABLE}
    TYPEABLE with type t = A.t list
implicit module Typeable_pair{A:TYPEABLE}{B:TYPEABLE} :
    TYPEABLE with type t = A.t * B.t
(* ... *)
```

#### Defining `typeable`: an initial attempt

We might try defining `typerep` as follows:

````ocaml
type _ typerep =
  Int : int typerep
| Bool : bool typerep
| List : 'a typerep -> 'a list typerep
| Option : 'a typerep -> 'a option typerep
| Pair : 'a typerep * 'b typerep -> ('a * 'b) typerep
````


which allows us to write a suitably-typed equality predicate:

````ocaml
let rec eqrep :
 type a b.a typerep -> b typerep -> (a,b) eql option =
 fun l r -> match l, r with
    Int, Int -> Some Refl
  | Bool, Bool -> Some Refl
  | List s, List t ->
     (match eqrep s t with
        Some Refl -> Some Refl
      | None -> None)
  | Option s, Option t ->
     (match eqrep s t with
        Some Refl -> Some Refl
      | None -> None)
  | Pair (s1, s2), Pair (t1, t2) ->
     (match eqrep s1 t1, eqrep s2 t2 with
        Some Refl, Some Refl -> Some Refl
      | _ -> None)
  | _ -> None
````

````ocaml
let int = Int
let bool = Bool
let list t = List t
let option t = Option t
let ( * ) l r = Pair (l, r)
````


We can use `eqrep` to compare type representations for equality:

````ocaml
let _ = eqrep int int
````

````ocaml
let _ = eqrep (int * bool) (bool * int)
````

````ocaml
let _ = eqrep (list int) (list int)
````


However, this definition is *closed*: it only supports the five
constructors in the definition, and cannot be extended.  Generic
functions are much more useful when they can be extended to support
arbitrary user-defined types.

### Extensible variants

In order to build an extensible definition of type equality we'll use OCaml's
extensible variants.  Extensible variants behave like standard OCaml variants,
except they can be extended with new constructors after they have been
defined.

Here's a **definition** for an extensible variant type `ty`:

````ocaml
type 'a ty = ..
````


We can **extend** `ty` with new constructors using the `+=` syntax:

````ocaml
type 'a ty +=
  A : int list -> int ty
| B : float list -> 'a ty
````


**Constructing** values of an extensible variant type looks just the same as
for a standard variant:

````ocaml
let _ = A [1;2;3]  (* No different to standard variants *)
````


Finally, **matching** for extensible variants looks similar to matching for
standard variants, but the only way to write an exhaustive match is to use a
wildcard or variable:

````ocaml
let f : type a. a ty -> string = function
  A _ -> "A"
| B _ -> "B"
| _ -> "unknown" (* All matches must be open *)
````


### Defining `typeable`, extensibly

We can use extensible variants to create an extensible type representation.
Here's a definition for an extensible variant type `type_rep`:

````ocaml
type _ type_rep = ..
````


We'll use `type_rep` to define `TYPEABLE`:

````ocaml
module type TYPEABLE = sig
  type t
  val type_rep : t type_rep Lazy.t
  val eqrep : 'other type_rep -> (t, 'other) eql option
end
````


```

(`Lazy.t` is used here to work around a limitation of recursive modules in OCaml.)

We must extend `type_rep` with one constructor for each type.  Here's a
constructor for `int`:

````ocaml
type _ type_rep += Int : int type_rep
````


and the `TYPEABLE` instance for `int`.  The `eqrep` which takes a type
representation and determines whether it is equal to `Int`:

````ocaml
implicit module Typeable_int
  : TYPEABLE with type t = int =
struct
  type t = int
  let type_rep = lazy Int
  let eqrep : type other. other type_rep -> (int, other) eql option =
   function
    | Int -> Some Refl
    | _ -> None
end
````


Each new instance of `TYPEABLE` requires a corresponding extension of
`type_rep`.  Here's a second extension of `type_rep` with a
constructor for `list`:

````ocaml
type _ type_rep += List : 'a type_rep -> 'a list type_rep
````


and here's the corresponding `TYPEABLE` instance:

````ocaml
implicit module Typeable_list{A:TYPEABLE}
  : TYPEABLE with type t = A.t list =
struct
  type t = A.t list
  let type_rep = lazy (List (Lazy.force A.type_rep))
  let eqrep : type other. other type_rep -> (A.t list, other) eql option =
   function
    | List b -> (match A.eqrep b with
                   Some Refl -> Some Refl
                 | None -> None)
    | _ -> None
end
````


The implementations for `option`, `bool`, etc. are similar and we can,
of course, extend `type_rep` and `TYPEABLE` with constructors and
instances for user-defined types.

### Additional `TYPEABLE` instances

Here are a few more instances of `TYPEABLE`, for `bool`, `string`, and `*`:

````ocaml
type _ type_rep += Bool : bool type_rep
implicit module Typeable_bool
  : TYPEABLE with type t = bool =
struct
  type t = bool
  let type_rep = lazy Bool
  let eqrep : type other. other type_rep -> (bool, other) eql option =
   function
    | Bool -> Some Refl
    | _ -> None
end
type _ type_rep += String : string type_rep
implicit module Typeable_string
  : TYPEABLE with type t = string =
struct
  type t = string
  let type_rep = lazy String
  let eqrep : type other. other type_rep -> (string, other) eql option =
   function
    | String -> Some Refl
    | _ -> None
end
type _ type_rep += Option : 'a type_rep -> 'a option type_rep
implicit module Typeable_option{A:TYPEABLE}
  : TYPEABLE with type t = A.t option =
struct
  type t = A.t option
  let type_rep = lazy (Option (Lazy.force A.type_rep))
  let eqrep : type other. other type_rep -> (A.t option, other) eql option =
   function
    | Option b ->
       (match A.eqrep b with
         | Some Refl -> Some Refl
         | _ -> None)
    | _ -> None
end
type _ type_rep += Pair : 'a type_rep * 'b type_rep -> ('a * 'b) type_rep
implicit module Typeable_pair{A:TYPEABLE}{B:TYPEABLE}
  : TYPEABLE with type t = A.t * B.t =
struct
  type t = A.t * B.t
  let type_rep = lazy (Pair (Lazy.force A.type_rep, Lazy.force B.type_rep))
  let eqrep : type other. other type_rep -> (A.t * B.t, other) eql option =
   function
    | Pair (c, d) ->
       (match A.eqrep c, B.eqrep d with
         | Some Refl, Some Refl -> Some Refl
         | _ -> None)
    | _ -> None
end
````



## Traversing datatypes

The second interface in our generic programming library supports
traversing values.  As with `TYPEABLE`, we'll start with a signature
(module type):

```
module type DATA
```

The `DATA` interface will expose two operations on values

* `gmapT` applies a function to every sub-node of a value, using the
  results to reconstruct the value.
* `gmapQ` applies a function to every sub-node of a value, collecting
  the results in a list.

Since the sub-nodes of a value can be of many different types the function
passed as argument to `gmapT` must be polymorphic.  We write the type of
`gmapT` as follows:

```
val gmapT : {D:DATA} -> ({B:DATA} -> B.t -> B.t) -> D.t -> D.t
```

The second implicit argument `{B:DATA}` appears below the
top-level, so `gmapT` depends on first-class (System F-style) polymorphism.

Similarly, `gmapQ` has the following type:

```
val gmapQ : {D:data} -> ({B:DATA}  -> B.t  ->  'u) -> D.t -> 'u list
```

It is worth noting that the type variable `'u` is implicitly
quantified at the top level: the `'u` in the return type of `gmapQ`
and in the return type of its function argument are the same. 

The `DATA` signature contains the functions `gmapT` and `gmapQ` along
with a `TYPEABLE` module and the name of the constructor used to build
a value.  Since the types of `gmapT` and `gmapQ` themselves involve
`DATA`, we need a recursive module to build the signature:

````ocaml
module rec R : sig
  type genericT = {B:R.DATA} -> B.t -> B.t
  type 'u genericQ = {B:R.DATA}  -> B.t  ->  'u
  module type DATA = sig
    type t
    module Typeable : TYPEABLE with type t = t
    val gmapT : genericT -> t -> t
    val gmapQ : 'u genericQ -> t -> 'u list
    val constructor_: t -> string
  end
end = R
include R
````


As usual, the top-level functions `gmapT`, `gmapQ`, and `constructor_`
simply project out the corresponding `DATA` members.  It is convenient
for them to take the implicit argument after the function argument.

````ocaml
let gmapT : genericT -> genericT =
    fun f {D:DATA} x -> D.gmapT f x
let gmapQ : 'u genericQ -> 'u list genericQ =
    fun f {D:DATA} x -> D.gmapQ f x
let constructor_: string genericQ =
    fun {D:DATA} x -> D.constructor_ x
````


### DATA instances

The `DATA` instances for primitive types (`int`, `bool`, `string`,
etc.) are straightforward, since primitive values have no subnodes.

Here is the `DATA` instance for `int`:

````ocaml
implicit module Data_int : DATA with type t = int = struct
  type t = int
  module Typeable = Typeable_int
  let gmapT f x = x
  let gmapQ f x = []
  let constructor_ = string_of_int
end
````


Since an `int` has no subnodes, both `gmapT` and `gmapQ` ignore their
function arguments.

Next, we'll look at the `DATA` instance for `list`.  A list value is
either `[]`, in which case there are no subnodes, or `x :: xs` for
some `x` and `xs`, in which case there are two subnodes --- `x` and
`xs`.

Here is the `DATA` instance for the `list` constructor.  The type
constructor `list` has one type parameter, and so the instance for
`list` takes one `DATA` instance as an argument.

Since the `list` type is recursive, the instance for `list` involves a
recursive module.  (It is likely that the syntax for recursive
instances will be simplified in a future OCaml release.)

````ocaml
implicit module Data_list {A: DATA} : DATA with type t = A.t list =
struct
  module rec R : DATA with type t = A.t list =
  struct
    type t = A.t list 
    module Typeable = Typeable_list{A.Typeable}
    let gmapT (f : genericT) (l : t) =
      match l with
      | [] -> []
      | x :: xs -> f x :: f {R} xs
    let gmapQ (q : _ genericQ) (l : t) =
      match l with
      | [] -> []
      | x :: xs -> [q x; q {R} xs]
    let constructor_ = function
      | [] -> "[]"
      | _::_ -> "::"
  end
  include R
end
````



### Additional `DATA` instances

Here are a few more instances of `DATA`, for `bool`, `string`,
`option`, and `*`:

````ocaml
implicit module Data_bool : DATA with type t = bool = struct
  type t = bool
  module Typeable = Typeable_bool
  let gmapT f x = x
  let gmapQ f x = []
  let constructor_ = string_of_bool
end
implicit module Data_string : DATA with type t = string = struct
  type t = string
  module Typeable = Typeable_string
  let gmapT f x = x
  let gmapQ f x = []
  let constructor_ = Printf.sprintf "%S"
end
implicit module Data_option {A: DATA} : DATA with type t = A.t option =
struct
  type t = A.t option 
  module Typeable = Typeable_option{A.Typeable}
  let gmapT (f : genericT) (o : t) =
    match o with None -> None | Some x -> Some (f x)
  let gmapQ (q : _ genericQ) (o : t) =
    match o with None -> [] | Some x -> [q x]
  let constructor_ = function
    | None -> "None"
    | Some _ -> "Some"
end
implicit module Data_pair {A: DATA} {B: DATA}
 : DATA with type t = A.t * B.t =
struct
  type t = A.t * B.t
  module Typeable = Typeable_pair{A.Typeable}{B.Typeable}
  let gmapT (f : genericT) ((x, y) : t) = (f x, f y)
  let gmapQ (q : _ genericQ) ((x, y) : t) = [q x; q y]
  let constructor_ _ = "(,)"
end
````


### Generic schemes

The functions `gmapT` and `gmapQ` provide access to the immediate subnodes of
a value.  We can use them to build functions which provide access to *all*
subnodes.

#### Generic maps

Here are two functions which use `gmapT` to map over every subnode of a value.
The `everywhere` function traverses a value, applying the argument function
bottom-up to every value encountered during the traversal.

````ocaml
let rec everywhere : genericT -> genericT =
 fun (f : genericT) {X:DATA} x ->
    f ((gmapT (everywhere f) : genericT) x)
````


The `everywhere'` function is like `everywhere`, but applies the function
top-down:

````ocaml
let rec everywhere' : genericT -> genericT =
  fun (f : genericT) {X:DATA} x ->
     (gmapT (everywhere' f) : genericT) (f {X} x)
````


Applying a function everywhere can be computationally expensive.  It is
sometimes convenient to stop traversing when some condition is satisfied.
Here is a function, `everywhereBut`, that takes an additional function which
is called to determine whether to stop the traversal:

````ocaml
let rec everywhereBut : bool genericQ -> genericT -> genericT =
  fun (stop : bool genericQ) (f : genericT) {X:DATA} x ->
    if stop x then x else f ((gmapT (everywhereBut stop f) : genericT) x)
````


### Using generic maps

It is sometimes useful to apply a function to every value of a
particular type within some larger value. For example, we might like
to increment all the integers in a list of pairs of booleans and
integers.

The following function, `mkT` (defined with an auxiliary function
`appT`) makes it easier to write such cases.  The implicit argument to
`mkT` is a type representation for the values of interest, and the the
non-implicit argument is a function to apply to those values.  The
result is a polymorphic function suitable for passing to `everywhere`.

````ocaml
let appT (type a) (type b) (module A : TYPEABLE with type t = a)
    (b : b type_rep) (g : b -> b) (x : a) : a =
    match A.eqrep b with
  | Some Refl -> g x
  | _         -> x
let mkT : {T:TYPEABLE} -> (T.t -> T.t) -> genericT =
  fun {T:TYPEABLE} g {D: DATA} ->
    appT (module D.Typeable) (Lazy.force T.type_rep) g
````


Here's `mkT` in action:

````ocaml
let _ = everywhere (mkT succ) [(false, 1); (false, 2); (true, 3)]
````


### Generic queries

The `everywhere` function extended `gmapT` to map over all subnodes.
Similarly, we can write a function `everything` which extends `gmapQ` to
query all subnodes:

````ocaml
let rec everything : 'r. ('r -> 'r -> 'r) -> 'r genericQ -> 'r genericQ =
  fun join g {X: DATA} x ->
    let f = g x in  List.fold_left join f (gmapQ (everything join g) x)
````



The type `'r` represents the type of the results of the query.  The
second argument combines two results to produce a single result.

We can also write a generic query function `everythingBut` which includes a
stop condition, analogous to `everywhereBut`:

````ocaml
let rec everythingBut : 'r. ('r -> 'r -> 'r) -> ('r * bool) genericQ -> 'r genericQ =
  fun (@) (stop : _ genericQ) {X: DATA} x ->
    match stop x with
    | v, true -> v
    | v, false -> List.fold_left (@) v (gmapQ (everythingBut (@) stop) x)
````


### Using generic queries

Finally, we can write a function which builds a polymorphic value
suitable for passing to `everything` and `everythingBut`.  The `mkQ`
function takes a typeable instance `T`, a "default" result argument
`u` and a function `g` which turns values of type `T.t` into results.
The result of `mkQ` is a polymorphic function that applies `g` to
arguments of type `t` and returns `u` otherwise:

````ocaml
let appQ (type a) (type b) (type u) (module A : TYPEABLE with type t = a)
    (b : b type_rep) (u:  u) (g : b -> u) (x: a) : u =
    match A.eqrep b with
  | Some Refl -> g x
  | _         -> u
let mkQ : 'u. {T:TYPEABLE} -> 'u -> (T.t -> 'u) -> 'u genericQ =
 fun {T:TYPEABLE} u g {D: DATA} x ->
   appQ (module D.Typeable) (Lazy.force T.type_rep) u g x
````


Here's `mkQ` in action, collecting every boolean value found in an
association list:

````ocaml
let _ = 
  everything (@) (mkQ {Typeable_list{Typeable_bool}} [] (fun x -> [x])) [(1, false); (2, true); (3, false)]
````


### Generic printing

We can use `gmapQ` and `gmapT` to write many functions besides generic maps
and queries.  Here is a generic printing function `gshow`, which can turn any
value that we can describe with `data` into a string:

````ocaml
let rec gshow {D:DATA} (v : D.t) =
  "("^ constructor_ v ^ String.concat " " (gmapQ gshow v) ^ ")"
````


Here's `gshow` in action:

````ocaml
let _ = gshow [1;2;3]
````

````ocaml
let _ = gshow [(1, true); (2, false)]
````


### Generic sizing

Here's another generic function, `gsize`, which counts the number of
nodes in a value:

````ocaml
let sum l = List.fold_left (+) 0 l

let rec gsize {D:DATA} (v : D.t) = 1 + sum (gmapQ gsize v)
````


An integer has no subnodes, so it has size 1.

````ocaml
let _ = gsize 3
````


A list of three integers has size 7 (three integers, three cons cells,
and nil).

````ocaml
let _ = gsize [1;2;3]
````


What's the size of a list of pairs?

````ocaml
let _ = gsize [(1,false); (2,false); (3,true)]
````


### Efficiency issues

Generic traversals are often rather slow, since they typically perform
many type comparisons and sometimes traverse subvalues unnecessarily.
The staging lectures show how to turn generic functions into code
generators that build much more efficient type-specialized functions.
