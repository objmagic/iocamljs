## Generic programming

Generic programming involves functions which operate over a wide range of data
types.  Unlike (parametrically) polymorphic functions, which operate uniformly
at all types, generic functions behave in a way varies predictably for each
type.

For any type `t` from variants, records and numeric types we can define a
variety of functions, including

* **equality**

  ```
  t -> t -> bool
  ```

* **ordering**

  ```
  t -> t -> int
  ```

* **hashing**

  ```
  t -> int
  ```

* **pretty-printing**

  ```
  t -> string
  ```

* **parsing**

  ```
  string -> t
  ```

* **serialising**

  ```
  t -> string
  ```

* **sizing**

  ```
  t -> int
  ```

OCaml provides a number of built-in generic functions, including functions for
equality, hashing, and serialisation.  Unfortunately, as we have seen, these
functions are not compatible with parametricity.  We can write generic
functions that behave parametrically by adding an additional parameter which
describes the *shape*.  For example, OCaml's built-in polymorphic equality has
the following type:

```
'a -> 'a -> bool
```

Our parametric equality function will have an additional argument of
type `'a data' which describes the shape of the values to be compared
for equality:

```
'a data -> 'a -> 'a -> bool
```

Most of this notebook is about one approach to defining and using the
`data` type.

## Type-indexed values

Values of the `data` type are an example of a *type-indexed value*.  The idea
of type-indexed values is to represent types by defining a parameterised type

```
type 'a t
```

together with one constructor value or function for each type constructor in
the program

```
val bool : bool t
val int : int t
val ( * ) : 'a t -> 'b t -> ('a * 'b) t 
val list : 'a t -> 'a list t 
val option : 'a t -> 'a option t 
(* etc. *)
```

and then using these constructors to build representations of type
expressions.  For example, we might represent the type

```
(int * bool) option list
```

using the expression:

```
list (option (int * bool))
```

Type-indexed values can be used to define a variety of generic
functions.  For example, we might define a type-indexed value that
allows us to compare values for equality:

````ocaml
type 'a t = 'a -> 'a -> bool

let bool : bool t =
  fun l r -> match l, r with
    false, false
  | true , true -> true
  | _           -> false

let ( * ) : 'a t -> 'b t -> ('a * 'b) t =
  fun leq req (l1, r1) (l2, r2) -> leq l1 l2 && req r1 r2

(* etc. *)
````


or for printing

````ocaml
type 'a t = 'a -> string

let bool : bool t =
  function
    false -> "false"
  | true -> "true"

let ( * ) : 'a t -> 'b t -> ('a * 'b) t =
  fun lpr rpr (l, r) -> "("^ lpr l ^", "^ rpr r ^")"

(* etc. *)
````


We're going to look at defining more generic type-indexed values that make it
possible to write a wide variety of generic functions in a concise way.  In
particular, we'll define:

* a type-indexed value for testing whether two types are equal
* a type-indexed value for accessing the immediate sub-nodes of a value.

These two type-indexed values will form the basis of a generic programming
library.

## Type equality

We'll start by building a type-indexed value for type equality ---
that is, for testing whether two types are equal.  We'll need a type,
which we'll call `typeable`, to represent types as concrete values:

```
type 'a typeable 
```

The `typeable` type should support one operation, `=~=`, which
compares two `typeable` values for equality, returnining a GADT as
proof if they're determined to be equal.

```
val (=~=) : 'a typeable ->
            'b typeable ->
              ('a,'b) eql option
```

The `eql` type is the familiar equality GADT:

````ocaml
type (_, _) eql = Refl : ('a, 'a) eql
````


Finally, we'll need a value of type `typeable` for each type constructor:

```  
module Typeable :
sig
  val int : int typeable
  val bool : bool typeable
  val list : 'a typeable -> 'a list typeable
  val ( * ) : 'a typeable -> 'b typeable ->
     ('a * 'b) typeable
  (* ... *)
end
```

#### Defining `typeable` an initial attempt

We might try defining `typeable` like this:

````ocaml
type _ typeable =
  Int : int typeable
| Bool : bool typeable
| List : 'a typeable -> 'a list typeable
| Option : 'a typeable -> 'a option typeable
| Pair : 'a typeable * 'b typeable -> ('a * 'b) typeable
````


which allows us to write a suitably-typed equality predicate:

````ocaml
let rec (=~=) :
 type a b.a typeable -> b typeable -> (a,b) eql option =
 fun l r -> match l, r with
    Int, Int -> Some Refl
  | Bool, Bool -> Some Refl
  | List s, List t ->
     (match s =~= t with
        Some Refl -> Some Refl
      | None -> None)
  | Option s, Option t ->
     (match s =~= t with
        Some Refl -> Some Refl
      | None -> None)
  | Pair (s1, s2), Pair (t1, t2) ->
     (match s1 =~= t1, s2 =~= t2 with
        Some Refl, Some Refl -> Some Refl
      | _ -> None)
  | _ -> None
````

````ocaml
let int = Int
let bool = Bool
let list t = List t
let option t = Option t
let ( * ) l r = Pair (l, r)
````


We can use `=~=` to compare type representations for equality:

````ocaml
int =~= int
````

````ocaml
(int * bool) =~= (bool * int)
````

````ocaml
list int =~= list int
````


However, this definition is *closed*: it only supports the five
constructors in the definition, and cannot be extended.  Generic
functions are much more useful when they can be extended to support
arbitrary user-defined types.

### Extensible variants

In order to build an extensibile definition of type equality we'll use OCaml's
extensible variants.  Extensible variants behave like standard OCaml variants,
except they can be extended with new constructors after they have been
defined.

Here's a **definition** for an extensible variant type `t`:

````ocaml
type 'a t = ..
````


We can **extend** `t` with new constructors using the `+=` syntax:

````ocaml
type 'a t +=
  A : int list -> int t
| B : float list -> 'a t
````


**Constructing** values of an extensible variant type looks just the same as
for a standard variant:

````ocaml
A [1;2;3]  (* No different to standard variants *)
````


Finally, **matching** for extensible variants looks similar to matching for
standard variants, but the only way to write an exhaustive match is to use a
wildcard or variable:

````ocaml
let f : type a. a t -> string = function
  A _ -> "A"
| B _ -> "B"
| _ -> "unknown" (* All matches must be open *)
````


### Defining `typeable`, extensibly

We can use extensible variants to create an extensible type representation.
Here's a definition for an extensible variant type `type_rep`:

````ocaml
type _ type_rep = ..
````


A `typeable` value is now a pair of a `type_rep` and an equality function
`eqty` which takes some other `type_rep` as an argument:

````ocaml
type 'a typeable = {
  type_rep : 'a type_rep;
  eqty : 'b. 'b type_rep -> ('a, 'b) eql option;
}
````


We must extend `type_rep` with one constructor for each type.  Here's a
constructor for `int`:

````ocaml
type _ type_rep += Int : int type_rep
````


and a definition of a function `eq_int` which takes a type representation and
determines whether it is equal to `Int`:

````ocaml
let eq_int :
 type b. b type_rep -> (int, b) eql option =
 function Int -> Some Refl | _ -> None
````


We can use `Int` and `eq_int` together to build an element of the
utype-indexed value:

````ocaml
let int = { type_rep = Int; eqty = eq_int }
````


Here's a second `type_rep` constructor for `list`:

````ocaml
type _ type_rep += List : 'a type_rep -> 'a list type_rep
````


and a function `eq_list`

````ocaml
let eq_list :
 type a b.a typeable -> b type_rep -> (a list,b) eql option =
 fun t -> function
     List a -> (match t.eqty a with
                 Some Refl -> Some Refl
               | None -> None)
   | _ -> None
````


such that `eq_list t (List t)` returns a proof of equality and all other calls
return `None`.

We can use `List` and `eq_list` to build a second element of the type-indexed
value:

````ocaml
let list a = { type_rep = List a.type_rep;
               eqty = fun t -> eq_list a t }
````


The implementations for `option`, `bool`, etc. are similar and we can, of
course, extend `type_rep` and `typeable` with constructors and definitions for
user-defined types.

### More comprehensive definitions for `typeable`

Here are some more comprehensive definitions for `typeable`.

````ocaml
module Typeable :
sig
  val int : int typeable
  val bool : bool typeable
  val float : float typeable
  val list : 'a typeable -> 'a list typeable
  val ( * ) : 'a typeable -> 'b typeable -> ('a * 'b) typeable
end =
struct
  type _ type_rep += Int : int type_rep
  let eq_int : type b. b type_rep -> (int, b) eql option =
    function Int -> Some Refl | _ -> None
  let int = { type_rep = Int; eqty = eq_int }

  type _ type_rep += Bool : bool type_rep
  let eq_bool : type b. b type_rep -> (bool, b) eql option =
    function Bool -> Some Refl | _ -> None
  let bool = { type_rep = Bool; eqty = eq_bool }

  type _ type_rep += Float : float type_rep
  let eq_float : type b. b type_rep -> (float, b) eql option =
    function Float -> Some Refl | _ -> None
  let float = { type_rep = Float; eqty = eq_float }

  type _ type_rep += List : 'a type_rep -> 'a list type_rep
  let eq_list : type a b. a typeable -> b type_rep -> (a list, b) eql option =
    fun tya -> function
        List a ->
        begin match tya.eqty a with
            Some Refl -> Some Refl
          | None -> None
        end
      | _ -> None
  let list a = { type_rep = List a.type_rep; eqty = fun t -> eq_list a t }

  type _ type_rep += Pair : 'a type_rep * 'b type_rep -> ('a * 'b) type_rep
  let eq_pair : type a b c. a typeable -> b typeable -> c type_rep -> (a * b, c) eql option =
    fun tya tyb -> function
        Pair (a, b) ->
        begin match tya.eqty a, tyb.eqty b with
            Some Refl, Some Refl -> Some Refl
          | _ -> None
        end
      | _ -> None
  let ( * ) a b =
    { type_rep = Pair (a.type_rep, b.type_rep);
      eqty = fun t -> eq_pair a b t }
end
````


### Implementing type equality, extensibly

Here is a definition of `=~=` using the extensible `typeable` definition:

````ocaml
let (=~=) :
 type a b.a typeable -> b typeable -> (a,b) eql option =
  fun a b -> a.eqty b.type_rep
````


As before, we can use `=~=` to compare type representations:

````ocaml
Typeable.(int =~= int)
````

````ocaml
Typeable.((int * bool) =~= (bool * int))
````

````ocaml
Typeable.(list int =~= list int)
````


## Traversing datatypes

The second type-indexed value in our generic programming library supports
traversing values.  As before, we'll start with a type for representing our
type descriptions:

```
type 'a data 
```

The `data` type will support two operations

* `gmapT`, which applies a function to every sub-node of a value, using the
  results to reconstruct the value:
* `gmapQ`, which applies a function to every sub-node of a value, collecting
  the results in a list.

Since the sub-nodes of a value can be of many different types the function
passed as argument to `gmapT` must be polymorphic.  We might write the type of
`gmapT` as follows:

```
val gmapT : 'a data -> ('b.'b data -> 'b -> 'b) -> 'a -> 'a
```

except that OCaml doesn't support first-class (System F-style) polymorphism,
so we must wrap the argument in a record:

```
type gmapT_arg = { f : 'b. 'b data -> 'b -> 'b; }

val gmapT : 'a data -> gmapT_arg -> 'a -> 'a
```

Similarly, `gmapQ` could have the following type:

```
val gmapQ : 'a data -> ('b.'b data  ->  'b  ->  'u) -> 'a -> 'u list
```

but the lack of polymorphic function arguments again means that we must use a
record for the argument type:

```
type 'u gmapQ_arg = { q : 'b. 'b term -> 'b -> 'u }
  
val gmapQ : 'a data -> 'u gmapQ_arg -> 'a -> 'u list
```

We'll make `data` a record containing the functions `gmapT` and `gmapQ` along
with a `typeable` value and a function for retrieving the name of the
constructor used to build a value:

````ocaml
type 'a data = {
  typeable : 'a typeable;
  gmapT : gmapT_arg -> 'a -> 'a;
  gmapQ : 'u. 'u gmapQ_arg -> 'a -> 'u list;
  ctor : 'a -> string;
}
and gmapT_arg = { f : 'b. 'b data -> 'b -> 'b; }
and 'u gmapQ_arg = { q : 'b. 'b data -> 'b -> 'u }
````


In addition to `data` and its operations `gmapT` and `gmapQ`, we'll need
constructors following the usual scheme.  Let's define the constructor `int`.

An integer has no sub-nodes, so both `gmapT` and `gmapQ` can ignore their
function arguments:

````ocaml
let gmapT_int f x = x
````

````ocaml
let gmapQ_int q x = []
````


We can use `gmapT_int` and `gmapQ_int` along with the typeable value for `int`
to build the constructor:

````ocaml
let int =
{ typeable = Typeable.int;
  gmapT = gmapT_int;
  gmapQ = gmapQ_int;
  ctor = string_of_int }
````


We can also define a constructor for `list`.  A list value is either `[]`, in
which case there are no subnodes, or `x :: xs` for some `x` and `xs`, in which
case there are two subnodes --- `x` and `xs`:

Here's a definition for the `list` constructor:

````ocaml
  let list : type a. a data -> a list data =
   fun a -> 
     let rec l = {
      typeable = Typeable.(list a.typeable);
      gmapT = (fun {f} -> function
                [] -> []
              | x :: xs -> f a x :: f l xs);
      gmapQ = (fun {q} -> function
                 | [] -> []
                 | x :: xs -> [q a x; q l xs]);
      ctor = (function [] -> "[]" | _::_ -> "::");
     } in l
````


### More comprehensive definitions for `data`

Here are some more comprehensive definitions for  `data`.

````ocaml
module Data :
sig
  val int : int data
  val bool : bool data
  val float : float data
  val list : 'a data -> 'a list data
  val ( * ) : 'a data -> 'b data -> ('a * 'b) data
end =
struct
  let primitive typeable show = {
    typeable;
    gmapT = (fun _ x -> x);
    gmapQ = (fun _ _ -> []);
    ctor = show;
  }

  let int = primitive Typeable.int string_of_int
  let bool = primitive Typeable.bool string_of_bool
  let float = primitive Typeable.float string_of_float

  let list : type a. a data -> a list data =
   fun a -> 
     let rec l = {
      typeable = Typeable.(list a.typeable);
      gmapT = (fun {f} -> function
                [] -> []
              | x :: xs -> f a x :: f l xs);
      gmapQ = (fun {q} -> function
                 | [] -> []
                 | x :: xs -> [q a x; q l xs]);
      ctor = (function [] -> "[]" | _::_ -> "::");
     } in l

  let ( * ) : type a b. a data -> b data -> (a * b) data = 
   fun a b -> {
    typeable = Typeable.(a.typeable * b.typeable);
    gmapT = (fun { f } (x, y) -> (f a x, f b y));
    gmapQ = (fun { q } (x, y) -> [q a x; q b y]);
    ctor = (fun _ -> "(,)");
   }
end
````

>

````ocaml
open Data
````


### Generic schemes

The functions `gmapT` and `gmapQ` provide access to the immediate subnodes of
a value.  We can use them to build functions which provide access to *all*
subnodes.

#### Generic maps

Here are two functions which use `gmapT` to map over every subnode of a value.
The `everywhere` function traverses a value, applying the argument function
bottom-up to every value encountered during the traversal.

````ocaml
let rec everywhere : 'a.'a data -> gmapT_arg -> 'a -> 'a =
  fun t f x -> f.f t (t.gmapT { f = fun t -> everywhere t f } x)
````


The `everywhere'` function is like `everywhere`, but applies the function
top-down:

````ocaml
let rec everywhere' : 'a.'a data -> gmapT_arg -> 'a -> 'a =
  fun t f x -> (t.gmapT { f = fun t y -> everywhere'  t  f  (f.f t y) } x)
````


Applying a function everywhere can be computationally expensive.  It is
sometimes convenient to stop traversing when some condition is satisfied.
Here is a function, `everywhereBut`, that takes an additional function which
is called to determine whether to stop the traversal:

````ocaml
let rec everywhereBut : 'a. 'a data -> bool gmapQ_arg -> gmapT_arg -> 'a -> 'a =
 fun t q f x -> 
  if q.q t x then x
  else f.f t (t.gmapT {f = fun t -> everywhereBut t q f} x)
````


### Using generic maps

It is sometimes useful to apply a function to every value of a particular type
within some larger value. For example, we might like to increment all the
integers in a list of pairs of booleans and integers.

The following function, `mkT`, makes it easier to write such cases.  The two
arguments to `mkT` are a type representation for the values of interest and a
function to apply to those values.  The result is a polymorphic function
suitable for passing to `everywhere`.

````ocaml
let mkT : type b. b typeable -> (b -> b) -> gmapT_arg =
  fun b g -> 
    let f : type a. a data -> a -> a =
      fun a x -> match a.typeable =~= b with
      | Some Refl -> g x
      | _         -> x
    in { f }
````


Here's `mkT` in action:

````ocaml
everywhere
  (list (bool * int))
  (mkT Typeable.int succ)
  [(false, 1); (false, 2); (true, 3)]
````


### Generic queries

The `everywhere` function extended `gmapT` to map over all subnodes.
Similarly, we can write a function `everything` which extends `gmapQ` to
query all subnodes:

````ocaml
let rec everything : 'a 'r. 'a data -> ('r -> 'r -> 'r) -> 'r gmapQ_arg -> 'a -> 'r =
  fun t k q x -> 
    List.fold_left k (q.q t x)
    (t.gmapQ {q = fun t -> everything t k q} x)
````


The `everything` function invoves two types --- `'a`, the type of values to
query, and `'r`, the type of results.  The second argument combines two
results to produce a single result.

We can also write a generic query fucntion `everythingBut` which includes a
stop condition, analogous to `everywhereBut`:

````ocaml
let rec everythingBut : 'a 'r. 'a data -> ('r -> 'r -> 'r) -> ('r * bool) gmapQ_arg -> 'a -> 'r =
  fun t k q x -> 
   let (v, stop) = q.q t x in
   if stop then v
   else List.fold_left k v (t.gmapQ {q = fun t -> everythingBut t k q} x)
````


### Using generic queries

Finally, we can write a function which builds a polymorphic value suitable for
passing to `everything` and `everythingBut`.  The `mkQ` function takes a type
representation `t` for some type `b`, a "default" result argument `r` and a
function `br` which turns values of type `b` into results.  The result of
`mkQ` is a polymorphic function which applies `br` to arguments of type `b`
and returns `r` otherwise:

````ocaml
let mkQ : type b r.b typeable  ->  r  ->  (b  ->  r)  ->  r gmapQ_arg =
 fun t r br -> 
  let q : type a. a data -> a -> r =
   fun a x -> 
    match a.typeable =~= t with
     Some Refl -> br x
   | None -> r
  in { q }
````


Here's `mkQ` in action, collecting every boolean value found in an association
list:

````ocaml
everything (list (int * bool))
  (@) (mkQ Typeable.bool [] (fun x -> [x]))
   [(1, false); (2, true); (3, false)]
````


### Generic printing

We can use `gmapQ` and `gmapT` to write many functions besides generic maps
and queries.  Here is a generic printing function `gshow`, which can turn any
value that we can describe with `data` into a string:

````ocaml
let rec gshow : 'a. 'a data -> 'a -> string =
  fun t v -> 
    "("^ t.ctor v
    ^ String.concat " " (t.gmapQ {q = gshow} v)
    ^ ")"
````


Here's `gshow` in action:

````ocaml
gshow (list (int * bool)) [(1, true); (2, false)]
````

````ocaml
gshow (list (list int)) [[1;2]; [4;5]]
````


### Generic sizing

Here's another generic size function which counts the number of nodes in a
value:

````ocaml
let sum = List.fold_left (+) 0
````

>

````ocaml
let rec gsize : 'a. 'a data -> 'a -> int =
  fun t v -> 1 + sum (t.gmapQ {q = gsize} v)
````


An integer has no subnodes, so it has size 1.

````ocaml
gsize int 3
````


A list of three integers has size 7 (three integers, three cons cells, and nil).

````ocaml
gsize (list int) [1;2;3]
````


What's the size of a list of pairs?

````ocaml
gsize (list (int * bool))
   [(1,false); (2,false); (3,true)]
````


### Efficiency issues

Generic traversals are often rather slow, since they typically perform many
type comparisons and sometimes traverse subvalues unnecessarily.  The staging
lectures show how to turn generic functions into code generators that build
much more efficient type-specialized functions.
