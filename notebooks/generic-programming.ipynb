```ocaml
type (_, _) eql = Refl : ('a, 'a) eql
```

```ocaml
type _ type_rep = ..
```

```ocaml
type 't typeable = {
  type_rep : 't type_rep;
  eqty : 's. 's type_rep -> ('t, 's) eql option;
}
```

```ocaml
let (=~~=) a b = a.eqty b.type_rep
```

```ocaml
type   genericT = { t: 't. 't data -> 't -> 't }
and 'u genericQ = { q: 't. 't data -> 't -> 'u }
and 't data = {
  typeable: 't typeable;
  gmapT : genericT -> 't -> 't;
  gmapQ : 'u. 'u genericQ -> 't -> 'u list;
  constructor: 't -> string
}
```

```ocaml
let gmapT f = { t = fun {gmapT} -> gmapT f }
let gmapQ f = { q = fun {gmapQ} -> gmapQ f }
let constructor = { q = fun {constructor} -> constructor }
```

```ocaml
let mkT : type t. t typeable -> (t -> t) -> genericT =
  fun t g ->
    { t = fun (type t') (d : t' data) (x : t') ->
          match d.typeable.eqty t.type_rep with
          | Some Refl -> (g x : t')
          | _         -> x }
```

```ocaml
let mkQ : type t u. t typeable -> u -> (t -> u) -> u genericQ =
  fun t u g ->
    { q = fun (type t') (d : t' data) (x : t') ->
          match d.typeable.eqty t.type_rep with
          | Some Refl -> (g x : u)
          | _         -> u }
```

```ocaml
type _ type_rep += List : 'a type_rep -> 'a list type_rep
type _ type_rep += Option : 'a type_rep -> 'a option type_rep
type _ type_rep += Pair : 'a type_rep * 'b type_rep -> ('a * 'b) type_rep
```

```ocaml
let typeable0_make : type a. unit -> a typeable =
  fun () ->
    let module M = struct
      type _ type_rep += T : a type_rep
      type t = a
      let eqty : type b. b type_rep -> (t, b) eql option =
        function T -> Some Refl | _ -> None
      let type_rep = T
    end in M.{ type_rep; eqty }
```

```ocaml
let typeable_int = typeable0_make ()
let typeable_bool = typeable0_make ()
let typeable_float = typeable0_make ()
let typeable_string = typeable0_make ()
```

```ocaml
let typeable_pair (type a) (type b) (a : a typeable) (b : b typeable) =
  let eqty : type c. c type_rep -> (a * b, c) eql option = function
      Pair (a', b') ->
      begin match a.eqty a', b.eqty b' with
          Some Refl, Some Refl -> Some Refl
        | _ -> None
      end
    | _ -> None

  and type_rep = Pair (a.type_rep, b.type_rep) in
  { eqty; type_rep }
```

```ocaml
let typeable_list (type a) (a : a typeable) =
  let eqty : type b. b type_rep -> (a list, b) eql option = function
    | List a' ->
      begin match a.eqty a' with
          Some Refl -> Some Refl
        | None -> None
      end
    | _ -> None
  and type_rep = List a.type_rep in
  { eqty; type_rep }
```

```ocaml
let typeable_option (type a) (a : a typeable) =
  let eqty : type b. b type_rep -> (a option, b) eql option = function
    | Option a' ->
      begin match a.eqty a' with
          Some Refl -> Some Refl
        | None -> None
      end
    | _ -> None
  and type_rep = Option a.type_rep in
  { eqty; type_rep }
```

```ocaml
let primitive typeable constructor =
  let gmapT _ x = x
  and gmapQ _ _ = [] in
  { typeable; constructor; gmapT; gmapQ }
```

```ocaml
let int = primitive typeable_int
    (fun c -> (string_of_int c))
```

```ocaml
let bool = primitive typeable_bool
    (fun c -> (string_of_bool c))
```

```ocaml
let float = primitive typeable_float
    (fun c -> (string_of_float c))
```

```ocaml
let string = primitive typeable_string
    (fun c -> (Printf.sprintf "%S" c))
```

```ocaml
let list a =
  let rec r =
    let typeable = typeable_list a.typeable
    and gmapT {t=f} l =
      match l with
        [] -> []
      | x :: xs -> f a x :: f r xs
                     
    and gmapQ {q} l =
      match l with
        [] -> []
      | x :: xs -> [q a x; q r xs]
                   
    and constructor = function
        [] -> "[]"
      | _::_ -> "::"
    in
    { gmapT; gmapQ; typeable; constructor }
  in r
```

```ocaml
let ( * ) a b =
  let typeable = typeable_pair a.typeable b.typeable
  and gmapT {t=f} (x, y) = (f a x, f b y)
  and gmapQ {q} (x, y) = [q a x; q b y]
  and constructor _ = "(,)" in
    { gmapT; gmapQ; typeable; constructor }
```

```ocaml
let option a =
  let typeable = typeable_option a.typeable
  and gmapT {t=f} o =
    match o with None -> None | Some x -> Some (f a x)
  and gmapQ {q} o =
    match o with None -> [] | Some x -> [q a x]
  and constructor = function
      None -> "None"
    | Some _ -> "Some" in
  { gmapT; gmapQ; typeable; constructor }
```

```ocaml
let sum l = List.fold_left (+) 0 l
```

```ocaml
let rec everywhere : genericT -> genericT =
  fun f ->
    { t = fun data x -> f.t data ((gmapT (everywhere f)).t data x) }
```

```ocaml
let rec everywhere' : genericT -> genericT =
  fun f ->
    { t =
        fun data x -> (gmapT (everywhere' f)).t data (f.t data x) }
```

```ocaml
let rec everywhereBut : bool genericQ -> genericT -> genericT =
  fun stop f ->
    { t =
        fun data x ->
          if stop.q data x then x
          else f.t data ((gmapT (everywhereBut stop f)).t data x) }
```

```ocaml
let rec everything : 'r. ('r -> 'r -> 'r) -> 'r genericQ -> 'r genericQ =
  fun h g ->
    { q =
        fun data x ->
          let f = g.q data x in
          List.fold_left h f
            ((gmapQ (everything h g)).q data x) }
```

```ocaml
let rec everythingBut : 'r. ('r -> 'r -> 'r) -> ('r * bool) genericQ -> 'r genericQ =
  fun h (stop : _ genericQ) ->
    { q =
        fun data x ->
          match stop.q data x with
          | v, true -> v
          | v, false -> List.fold_left h v
                          ((gmapQ (everythingBut h stop)).q data x) }
```

```ocaml
let rec gsize : int genericQ =
  { q = fun data v -> 1 + sum ((gmapQ gsize).q data v) }
```

```ocaml
let rec gshow =
  { q =
      fun data v ->
        "("^ constructor.q data v
        ^ String.concat " " ((gmapQ gshow).q data v)
        ^ ")" }
```
